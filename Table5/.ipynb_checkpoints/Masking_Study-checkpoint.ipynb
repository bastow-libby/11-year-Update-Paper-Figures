# Masking Study Comparing 1D and 2D Fits

The objective of this study is to reconcile one-dimensional and two-dimensional fits of the relative intensity skymaps for the IceTop eleven-year analysis. 

```python
%load_ext autoreload
%autoreload 2
```

```python
from glob import glob

import numpy as np
import healpy as hp
import matplotlib.pyplot as plt

from Fit2d import multipoleFit2d, returnRI, multipoleFit1D
```

## Load Skymaps

```python
# Load in maps

relint_files = {}
relerr_files = {}
root = '/data/ana/CosmicRay/Anisotropy/IceTop/ITpass2/output/sidereal_unblinded'

tiers = [i for i in range(1, 5)]

for tier in tiers:
    files = glob(f'{root}/tier{tier}/reconstruction/relintensityiter/combined*.fits.gz')
    relint_files[tier] = sorted(files)[-1]

    files = glob(f'{root}/tier{tier}/reconstruction/significanceiter/significance*.fits.gz')
    relerr_files[tier] = sorted(files)[-1]

print(f'Root directory: {root}')
print('Relative intensity files:')
for tier, map_file in relint_files.items():
    print(f'  Tier {tier} : {{root}}{map_file.replace(root,"")}')

print('Relative error files:')
for tier, map_file in relerr_files.items():
    print(f'  Tier {tier} : {{root}}{map_file.replace(root,"")}')
```

## Statement of the problem

With a standard IceTop masking of -35 degrees, we found a disagreement between the best-fit dipole amplitudes as a function of energy

```python
#nfiles = len(relint_files.keys())
#l = 3    # Fit up to octupole term
#decmin = -90.
#decmax = -35.

#amp, phase, amp_err, phase_err = np.zeros((4, nfiles))
#amp_1d, phase_1d, amp_err_1d, phase_err_1d = np.zeros((4, nfiles))
```

```python
# 1D fit requires a scaling factor based on FoV
def scaling_factor(decmin, decmax):
    dmin = decmin * np.pi/180.
    dmax = decmax * np.pi/180.
    return 2*(np.sin(dmin) - np.sin(dmax)) / (dmin - dmax + np.cos(dmin)*np.sin(dmin) - np.cos(dmax)*np.sin(dmax))
```

```python
# Calculate 1D fit with standard masking
def get_1d_fit_vals(relint_files, relerr_files, decmin, decmax):

    tiers = sorted(relint_files.keys())
    amp, phase, amp_err, phase_err = np.zeros((4, len(tiers)))
    
    for i, tier in enumerate(tiers):
    
        data, bg, relint = hp.read_map(relint_files[tier], range(3))
        relerr,_,_ = hp.read_map(relerr_files[tier], range(3))
        
        ra, ri, sigmax, sigmay = returnRI(relint, relerr, nbins=24, decmin=decmin, decmax=decmax)
        popt, pcov, chi2 = multipoleFit1D(ra, ri, 3, sigmay)
        scale = scaling_factor(decmin, decmax)
    
        amp[i]   = popt[0] * scale
        phase[i] = popt[1]
    
        # Amplitude/phase correction for negative signs
        if amp[i] < 0:
            amp[i] *= -1
            phase[i] += np.pi
        if phase[i] < 0:
            phase[i] += 2*np.pi
        phase[i] %= 2*np.pi
        
        err = np.sqrt(np.diag(pcov))
        amp_err[i]   = err[0] * scale
        phase_err[i] = err[1]

    return amp, phase, amp_err, phase_err
```

```python
# Calculate 2D fit with standard masking
def get_2d_fit_vals(relint_files, relerr_files, decmin, decmax):

    tiers = sorted(relint_files.keys())
    amp, phase, amp_err, phase_err = np.zeros((4, len(tiers)))

    for i, tier in enumerate(tiers):
    
        data, bg, relint = hp.read_map(relint_files[tier], range(3))
        relerr,_,_ = hp.read_map(relerr_files[tier], range(3))
        
        popt, perr, chi2, ndof, pvalue = multipoleFit2d(relint, relerr, 3, decmin=decmin, decmax=decmax)
    
        a = np.reshape(popt, (-1,2))
        amp[i], phase[i] = a[0]
    
        # Amplitude/phase correction for negative signs
        if amp[i] < 0:
            amp[i] *= -1
            phase[i] += np.pi
        if phase[i] < 0:
            phase[i] += 2*np.pi
        phase[i] %= 2*np.pi
        
        e = np.reshape(perr, (-1,2))
        amp_err[i], phase_err[i] = e[0]

    return amp, phase, amp_err, phase_err
```

```python
decmin = -90.
decmax = -35.

amp_1d, phase_1d, amp_err_1d, phase_err_1d = get_1d_fit_vals(relint_files, relerr_files, decmin, decmax)
amp_2d, phase_2d, amp_err_2d, phase_err_2d = get_2d_fit_vals(relint_files, relerr_files, decmin, decmax)
```

```python
fig, ax = plt.subplots()
common_opts = {'marker':'.', 'markersize':8, 'ls':'-'}
ax.errorbar(tiers, np.abs(amp_1d), yerr=amp_err_1d, label='1D fit values', **common_opts)
ax.errorbar(tiers, np.abs(amp_2d), yerr=amp_err_2d, label='2D fit values', **common_opts)
ax.legend(loc='lower right')
ax.set_xlabel('Tier')
ax.set_ylabel('Best-fit Dipole Amplitude')
plt.show()
```

## Testing fit behavior as a function of mask

We can test the sensitivity of each fit to the masking applied. Try it first by implementing a mask near the pole (altering `decmin`)

```python
decmin_values = np.linspace(-90., -70., 11)
decmax = -35.

tiers = sorted(relint_files.keys())
amp_1d = {tier:[] for tier in tiers}
amp_2d = {tier:[] for tier in tiers}

for decmin in decmin_values:

    amp_1d_i, _, _, _ = get_1d_fit_vals(relint_files, relerr_files, decmin, decmax)
    amp_2d_i, _, _, _ = get_2d_fit_vals(relint_files, relerr_files, decmin, decmax)

    for i, tier in enumerate(tiers):
        amp_1d[tier] += [amp_1d_i[i]]
        amp_2d[tier] += [amp_2d_i[i]]
```

```python
fig, ax = plt.subplots()
for i, tier in enumerate(tiers):
    points_1d = ax.plot(decmin_values, amp_1d[tier], '.', label=f'Tier {tier}')
    color = points_1d[0].get_color()
    points_2d = ax.plot(decmin_values, amp_2d[tier], 'x', color=color, label=f'_Tier {tier} (2d)')
    ax.legend()
plt.show()
```

Result: both fits appear to be pretty stable with changing masks near the pole. The fits disagree, but do not vary much in their values.

Next, test stability with respect to a changing `decmax`

```python
decmin = -90.
decmax_values = np.linspace(-35., -55., 9)

tiers = sorted(relint_files.keys())
amp_1d = {tier:[] for tier in tiers}
amp_2d = {tier:[] for tier in tiers}

for decmax in decmax_values:

    amp_1d_i, _, _, _ = get_1d_fit_vals(relint_files, relerr_files, decmin, decmax)
    amp_2d_i, _, _, _ = get_2d_fit_vals(relint_files, relerr_files, decmin, decmax)

    for i, tier in enumerate(tiers):
        amp_1d[tier] += [amp_1d_i[i]]
        amp_2d[tier] += [amp_2d_i[i]]
```

```python
fig, ax = plt.subplots()
for i, tier in enumerate(tiers):
    points_1d = ax.plot(decmax_values, amp_1d[tier], '.', label=f'Tier {tier}')
    color = points_1d[0].get_color()
    points_2d = ax.plot(decmax_values, amp_2d[tier], 'x', color=color, label=f'_Tier {tier} (2d)')
    ax.legend()
plt.show()
```

We can see better agreement between the two fits at a `decmax` of 40 degrees at all tiers. We could potentially go stricter, but want to avoid losing too much sky coverage. We can plot the values at a mask of 40 degrees to look for agreement within uncertainties

```python
decmin = -90.
decmax = -40.

amp_1d, phase_1d, amp_err_1d, phase_err_1d = get_1d_fit_vals(relint_files, relerr_files, decmin, decmax)
amp_2d, phase_2d, amp_err_2d, phase_err_2d = get_2d_fit_vals(relint_files, relerr_files, decmin, decmax)
```

```python
fig, ax = plt.subplots()
common_opts = {'marker':'.', 'markersize':8, 'ls':'-'}
ax.errorbar(tiers, np.abs(amp_1d), yerr=amp_err_1d, label='1D fit values', **common_opts)
ax.errorbar(tiers, np.abs(amp_2d), yerr=amp_err_2d, label='2D fit values', **common_opts)
ax.legend(loc='lower right')
ax.set_xlabel('Tier')
ax.set_ylabel('Best-fit Dipole Amplitude')
plt.show()
```

Note, however, that the 2D fit values ('x's on the masking study plots) do _not_ change significantly as a function of the masked declination values. It is therefore more stable than the 1D fit, and we are able to simply use the values from the 35 degree mask that our energy and angular resolution studies suggested:

```python
# Print values for the paper
decmin = -90.
decmax = -35.

amp_2d, phase_2d, amp_err_2d, phase_err_2d = get_2d_fit_vals(relint_files, relerr_files, decmin, decmax)
```

```python
for i, tier in enumerate(tiers):
    print(f'Tier {tier}:')
    print(f'  Amplitude (2D): {amp_2d[i]:.5f} ± {amp_err_2d[i]:.5f}')
    print(f'  Phase (2D):     {np.degrees(phase_2d[i]):.3f} ± {np.degrees(phase_err_2d[i]):.3f}')
```
