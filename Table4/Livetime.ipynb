```python
import numpy as np

from pathlib import Path
from collections import defaultdict
from datetime import datetime as dt

import grl_tools
```

```python
# Load Serap's icetop GRL files
grl_base = Path('/data/ana/CosmicRay/IceTop_GRL')
grl_files = sorted(grl_base.glob('IC86_20*.txt'))

it_goodruns = []
for grl_file in grl_files:
    it_goodruns += grl_tools.parse_seraps_GRL(str(grl_file))
```

```python
# Load information from i3live
start = '2011-05-13 00:00:00'
end   = '2022-05-13 00:00:00'
i3_goodruns = grl_tools.i3live_grl(start, end, icetop=True)
```

```python
# Calculate livetimes for icetop
i3_livetime = grl_tools.daily_livetime(i3_goodruns, it_goodruns)
```

```python
ic_year = 2011
start = dt(ic_year, 5, 13)
stop  = dt(ic_year+1, 5, 13)

tot_livetime = {ic_year: 0}

for day, run_dict in sorted(i3_livetime.items()):

    t = dt.strptime(day, '%Y-%m-%d')
        
    if t >= stop:
        ic_year += 1
        start = dt(ic_year, 5, 13)
        stop  = dt(ic_year+1, 5, 13)
        tot_livetime[ic_year] = 0

    tot_livetime[ic_year] += sum(run_dict.values())
```

```python
t_tot = 0
year_len_tot = 0

for year, t in tot_livetime.items():
    
    n_days = t/86400
    year_len = 366 if year in [2011,2015,2019] else 365
    duty_cycle = n_days / year_len

    t_tot += n_days
    year_len_tot += year_len
    
    print(f'Livetime for {year}: {n_days:.2f} d ({duty_cycle*100:.2f}%)')

print(f'Total livetime:   {t_tot:.2f} d ({t_tot/year_len_tot*100:.2f}%)')
```
